# **Part 1: Explanatory Data Analysis**

### **Descriptive Statistics**

Descriptive statistics are like a summary of a bunch of data. 

Imagine you have a large group of numbers or measurements—like the ages of everyone in your class or the scores from a test. 

Descriptive statistics help you understand what's going on with that data without looking at every single number individually.

Here’s how it works:

1. **Measures of Central Tendency:**

   - **Mean (Average):** You add up all the numbers and then divide by the total count of numbers. This tells you the average value.

   - **Median:** This is the middle value when you arrange all the numbers in order. If there’s an even number of values, you take the average of the two middle ones. It shows the center of your data.

   - **Mode:** This is the number that appears most frequently in your data. Sometimes there’s more than one mode if multiple numbers appear equally often.

2. **Measures of Spread (or Variability):**

   - **Range:** This is the difference between the highest and lowest values. It gives you an idea of how spread out the numbers are.

   - **Variance:** This measures how much the numbers differ from the mean (average). A high variance means the numbers are spread out; a low variance means they are close to the mean.

   - **Standard Deviation:** This is like the average distance of each number from the mean. It's a bit easier to interpret than variance because it’s in the same units as the original data.

3. **Shape of the Data:**

   - **Skewness:** This tells you if your data is symmetrical or if it leans more to one side. If most data is on the left, it’s positively skewed; if it’s on the right, it’s negatively skewed.
   
   - **Kurtosis:** This tells you how “peaked” or “flat” the data distribution is. High kurtosis means more data is in the tails (the ends), while low kurtosis means the data is more evenly spread out.

In summary, descriptive statistics help you get a quick overview of your data. They give you an idea of what’s typical (like the average), how much variation there is, and whether there’s anything unusual in the data.

1. **Count**: This represents the number of non-missing (valid) entries for each variable in the dataset. It tells us how many data points we have for each feature.

2. **Mean**: The average value of each variable. It gives us a central value for each feature.

3. **Standard Deviation (std)**: It measures how spread out the values are from the mean. A higher standard deviation indicates that the data points are more spread out.

4. **Min (Minimum)**: The smallest value in the dataset for each variable. It gives us the lower boundary of the data.

5. **25th Percentile (25%)**: Also known as the first quartile, it indicates that 25% of the data falls below this value. It’s a measure of the lower end of the data.

6. **50th Percentile (50%)**: Also known as the median, it’s the middle value of the dataset when ordered from lowest to highest. Half of the data points are below this value, and half are above.

7. **75th Percentile (75%)**: Also known as the third quartile, it indicates that 75% of the data falls below this value. It’s a measure of the upper end of the data.

8. **Max (Maximum)**: The largest value in the dataset for each variable. It gives us the upper boundary of the data.

These summary statistics provide a snapshot of the dataset, helping to quickly identify the central tendency, variability, and range of the data.

### **Histograms**

### 1. What are Histograms?

A histogram is a type of bar chart that represents the distribution of numerical data. 

It shows how often each range of values occurs by grouping the data into bins or intervals. 

Unlike a regular bar chart, which is used for categorical data, histograms are used for continuous data and help you see the shape of your data distribution.

### 2. What are the Components of a Histogram?

A histogram has several key components:

- **Bins/Intervals:** These are the ranges of values into which the data is grouped. Each bin covers a specific range of the data.

- **Bars:** Each bar represents the frequency (or count) of data points within a bin. The height of the bar shows how many data points fall into that particular bin.

- **X-Axis:** This axis represents the variable you are measuring. It’s divided into intervals or bins.

- **Y-Axis:** This axis represents the frequency or count of data points in each bin.

- **KDE Line (Kernel Density Estimate):** When you include `kde=True`, it adds a smooth curve over the histogram that estimates the probability density function of the variable. It helps you understand the distribution more clearly.

### 3. How Do I Interpret a Histogram?

When interpreting a histogram, you should look at several aspects:

- **Shape:** Look at the overall shape of the histogram. Is it symmetric, skewed to the left or right, or does it have multiple peaks? The shape can tell you a lot about the distribution of your data.

- **Center:** The center of the histogram shows where most of the data points are clustered. This gives you an idea of the average or typical value.

- **Spread:** The width of the histogram shows how spread out the data is. A wider histogram indicates more variability, while a narrower one shows less variability.

- **Outliers:** Look for bars that are far away from the rest of the data. These could represent outliers or unusual data points.

### **Boxplots**

Let's dive into boxplots and then create the visualizations.

### 1. What are Boxplots?

A boxplot, also known as a box-and-whisker plot, is a standardized way of displaying the distribution of data based on five summary statistics: 

the minimum, first quartile (Q1), median, third quartile (Q3), and maximum. 

Boxplots help you quickly visualize the spread and skewness of your data and identify any outliers.

### 2. What are the Components of a Boxplot?

A boxplot consists of several key components:

- **Box:** The box represents the interquartile range (IQR), which is the range between the first quartile (Q1, 25th percentile) and the third quartile (Q3, 75th percentile). This middle 50% of the data is shown by the box itself.

- **Median Line:** Inside the box, there is a line that represents the median (50th percentile) of the data. This gives you a sense of the data's center.

- **Whiskers:** These are the lines that extend from the box to the minimum and maximum values, excluding outliers. They show the range of the bulk of the data.

- **Outliers:** Outliers are individual points that lie beyond the whiskers. They represent data points that are unusually far from the rest of the data.

### 3. How Do I Interpret a Boxplot?

When interpreting a boxplot, you should consider:

- **Spread:** The width of the box shows the spread of the middle 50% of the data. A wider box indicates more variability, while a narrower box indicates less.

- **Skewness:** If the median line is not centered within the box, it suggests that the data is skewed. A box with a longer whisker on one side indicates skewness in that direction.

- **Outliers:** Points that fall outside the whiskers are considered potential outliers. These points are often worth investigating to see if they are errors, unusual values, or important data points.

### **ECDF Plots**

Let's go through ECDF plots step by step:

### 1. What are ECDF Plots?

An ECDF plot stands for Empirical Cumulative Distribution Function plot. It is a way to visualize the cumulative distribution of a dataset. 

The ECDF shows the proportion (or count) of observations that fall below or at each value in the dataset. 

Unlike histograms or boxplots, which show the distribution of data points in bins or ranges, ECDF plots provide a view of the entire dataset without any binning.

### 2. What are the Components of an ECDF Plot?

An ECDF plot consists of the following components:

- **X-Axis:** This axis represents the values of the variable.

- **Y-Axis:** This axis represents the cumulative proportion or count of data points that are less than or equal to each value on the X-axis.

- **Step Function Line:** The plot typically appears as a step function, with steps increasing as you move from left to right. Each step corresponds to a data point in the dataset.

### 3. How Do I Interpret an ECDF Plot?

Interpreting an ECDF plot involves looking at the following:

- **Cumulative Distribution:** The Y-axis shows how much of your data is below a certain value. For example, if a point at X=10 corresponds to Y=0.7, it means that 70% of the data points are less than or equal to 10.

- **Shape of the Curve:** A steep increase in the curve indicates that many data points have similar values, while a gradual slope indicates more spread out data.

- **Comparing Distributions:** ECDF plots are particularly useful for comparing distributions of different datasets or variables. You can overlay multiple ECDFs to see how different variables compare.

### 4. Provide ECDF Plots for Every Variable in the Dataset

Now, I'll generate ECDF plots for each variable in your dataset using `sns.ecdfplot` with the specified parameters.

```python
# Create a grid of plots for ECDF plots
fig, axes = plt.subplots(nrows=(len(data.columns) + n_cols - 1) // n_cols, ncols=n_cols, figsize=(10, 10))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Plot ECDFs for each variable
for i, col in enumerate(data.columns):
    sns.ecdfplot(data[col], color='black', ax=axes[i])
    axes[i].set_title(col)
    axes[i].set_xlabel('')
    axes[i].set_ylabel('')

# Remove any empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

### **QQ-Plot**

Let's explore QQ plots and then create the visualizations you requested.

### 1. What are QQ Plots?

A QQ plot, or Quantile-Quantile plot, is a graphical tool used to assess if a dataset follows a particular distribution, most commonly a normal distribution. 

It compares the quantiles of the data with the quantiles of a theoretical distribution. If the data closely follows the theoretical distribution, the points on the plot will fall along a straight line.

### 2. What are the Components of a QQ Plot?

A QQ plot has several key components:

- **X-Axis (Theoretical Quantiles):** This axis represents the expected quantiles from the theoretical distribution (like a normal distribution).

- **Y-Axis (Sample Quantiles):** This axis represents the actual quantiles from your data.

- **QQ Line (Reference Line):** Often, a reference line (usually at a 45-degree angle) is added to the plot. This line helps you see how closely the data matches the theoretical distribution. Points close to this line indicate a good fit.

### 3. How is a QQ Plot Created?

Creating a QQ plot involves a few simple steps:

1. **Sort the Data:** The data is sorted from smallest to largest.

2. **Calculate Quantiles:** The quantiles of the data are calculated.

3. **Theoretical Quantiles:** Quantiles from the theoretical distribution are calculated.

4. **Plotting:** The sample quantiles are plotted against the theoretical quantiles. If the data follows the theoretical distribution closely, the points will align along a straight line.

### 4. How Do I Interpret a QQ Plot?

Interpreting a QQ plot involves looking at how closely the points follow the reference line:

- **Straight Line:** If the points closely follow the straight line, the data likely follows the theoretical distribution (e.g., normal distribution).

- **Deviations:** If the points deviate from the line, it suggests that the data does not follow the theoretical distribution. 

  - **S-Shape:** An S-shaped curve indicates that the data might have heavier tails than the theoretical distribution.

  - **Upward/Downward Curvature:** If the points curve upwards or downwards, it might suggest skewness in the data.

### 5. Provide QQ Plots for Every Variable in the Dataset

Now, I'll generate QQ plots for each variable in your dataset using the `stats.probplot` function with a normal distribution as the reference.


```python
import scipy.stats as stats

# Create a grid of plots for QQ plots
fig, axes = plt.subplots(nrows=(len(data.columns) + n_cols - 1) // n_cols, ncols=n_cols, figsize=(10, 10))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Plot QQ plots for each variable
for i, col in enumerate(data.columns):
    stats.probplot(data[col].dropna(), dist="norm", plot=axes[i])
    axes[i].set_title(col)
    axes[i].get_lines()[1].set_color('red')  # Set the line color to red for visibility

# Remove any empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```


Here are the QQ plots for each variable in your dataset. 

Each plot compares the quantiles of your data to the quantiles of a normal distribution. 

You can use these plots to assess how closely each variable follows a normal distribution:

- **Straight Line:** If the points align closely with the red reference line, the variable likely follows a normal distribution.

- **Deviations:** If the points deviate significantly from the line, the variable may not be normally distributed. The nature of these deviations can indicate skewness or heavy tails in the data.

These plots are useful for understanding the distribution characteristics of your dataset's variables.

# **Part 2: Multiple Linear Regression Model**

### What is Multiple Linear Regression?

Multiple linear regression is a statistical technique used to model the relationship between one dependent variable (let's call it $ y $) and two or more independent variables (denoted as $ x_{ij} $). 

The goal is to find a linear equation that best predicts the value of $ y $ based on the values of the independent variables.

### Independent Variables $ x_{ij} $

- **$ i $**: Represents the index of the observations (or data points). If you have $ n $ observations, then $ i $ ranges from 1 to $ n $.

- **$ j $**: Represents the index of the independent variables (or features). If you have $ p $ independent variables, then $ j $ ranges from 1 to $ p $.

For example, if you have 100 observations (data points) and 3 independent variables, $ i $ ranges from 1 to 100, and $ j $ ranges from 1 to 3.

### Observations and Predictions

- **Observations**: These are the actual data points you collect in your dataset. Each observation includes a value for the dependent variable $ y_i $ and corresponding values for the independent variables $ x_{i1}, x_{i2}, \dots, x_{ip} $.
  
- **Predictions**: These are the values of $ y $ that the model predicts based on the independent variables. The predicted value for an observation $ i $ is denoted as $ \hat{y}_i $.

### The Multiple Linear Regression Model

The general form of the multiple linear regression equation is:

$
y_i = \beta_0 + \beta_1 x_{i1} + \beta_2 x_{i2} + \dots + \beta_p x_{ip} + \epsilon_i
$

Where:

- $ y_i $ is the dependent variable for observation $ i $.

- $ x_{ij} $ is the value of the $ j $-th independent variable for observation $ i $.

- $ \beta_0 $ is the intercept (the value of $ y $ when all $ x_{ij} = 0 $).

- $ \beta_j $ (for $ j = 1, 2, \dots, p $) are the coefficients that measure the impact of each independent variable on $ y $.

- $ \epsilon_i $ is the error term, representing the difference between the actual $ y_i $ and the predicted $ \hat{y}_i $.

### Assumptions of a Classical Linear Regression Model

For the multiple linear regression model to provide reliable estimates, certain assumptions must be met:

1. **Linearity**: The relationship between the dependent variable $ y $ and the independent variables $ x_{ij} $ should be linear. 

This means that the effect of each independent variable on $ y $ is consistent and additive.
   
   $
   y_i = \beta_0 + \beta_1 x_{i1} + \beta_2 x_{i2} + \dots + \beta_p x_{ip} + \epsilon_i
   $

2. **Independence of Errors**: The error terms $ \epsilon_i $ should be independent of each other. 

This means that the error for one observation should not be correlated with the error for another observation.

   $
   \text{Cov}(\epsilon_i, \epsilon_k) = 0 \quad \text{for } i \neq k
   $

3. **Homoscedasticity**: The variance of the error terms $ \epsilon_i $ should be constant across all levels of the independent variables. 

This means that the spread of the residuals (the differences between the observed and predicted values) should be consistent.

   $
   \text{Var}(\epsilon_i) = \sigma^2 \quad \text{for all } i
   $

4. **No Perfect Multicollinearity**: The independent variables should not be perfectly correlated with each other. 

This means that no independent variable can be exactly predicted by a combination of other independent variables.

   $
   \text{If } x_{ij} = a_1 x_{i1} + a_2 x_{i2} + \dots + a_{j-1} x_{i(j-1)}, \text{ then } \beta_j \text{ cannot be estimated}
   $

5. **Normality of Errors**: The error terms $ \epsilon_i $ should be normally distributed. 

This is particularly important for making statistical inferences (e.g., confidence intervals, hypothesis tests).

   $
   \epsilon_i \sim N(0, \sigma^2)
   $

### Summary of Assumptions in Mathematical Form

1. **Linearity**:
   $
   y_i = \beta_0 + \sum_{j=1}^{p} \beta_j x_{ij} + \epsilon_i
   $

2. **Independence of Errors**:
   $
   \text{Cov}(\epsilon_i, \epsilon_k) = 0 \quad \text{for } i \neq k
   $

3. **Homoscedasticity**:
   $
   \text{Var}(\epsilon_i) = \sigma^2 \quad \text{for all } i
   $

4. **No Perfect Multicollinearity**:
   $
   \text{No perfect linear relationship among the } x_{ij}
   $

5. **Normality of Errors**:
   $
   \epsilon_i \sim N(0, \sigma^2)
   $

These assumptions ensure that the multiple linear regression model is properly specified, allowing for valid predictions and inferences.

# **Part 3: Regression Diagnostics**

Regression diagnostics refer to the process of evaluating the validity and quality of a regression model. 

After fitting a regression model, it’s important to check whether the underlying assumptions of the model hold true, and to identify any potential issues that could affect the accuracy and reliability of the model's predictions.

In simple terms, regression diagnostics involve examining the model's performance and the behavior of the residuals (the differences between the observed and predicted values) to ensure that the model is correctly specified. 

This process helps identify problems such as unusual data points (like outliers), incorrect assumptions about the data, or the influence of certain observations that might distort the results.

By conducting regression diagnostics, you can assess whether the model is appropriate for the data, whether it provides meaningful predictions, and whether the conclusions drawn from the model are trustworthy. 

It’s a crucial step in the modeling process to ensure that the model accurately reflects the relationships in the data and that any decisions based on the model are well-founded.

```python
import statsmodels.api as sm

# Define the dependent variable
y = data['stem_diameter']

# Define the independent variables (excluding the dependent variable itself)
X = data.drop(columns=['stem_diameter'])

# Add a constant (intercept) to the model
X = sm.add_constant(X)

# Build the OLS regression model
model = sm.OLS(y, X)
ols_model = model.fit()
```

The OLS regression model has been successfully built using **stem_diameter** as the dependent variable and all remaining variables as independent variables.

## **Outliers**

### Understanding Outliers

**Outliers** are data points that differ significantly from other observations in the dataset. 

They can arise due to variability in the data, measurement errors, or other anomalies. 

Outliers can have a substantial impact on statistical analyses, including regression models, as they can distort the results, leading to misleading conclusions. 

In the context of regression, outliers can affect the estimates of the coefficients and reduce the model's accuracy.

### Understanding Studentized Residuals

**Studentized residuals** are a form of residuals that have been standardized by an estimate of their standard deviation, making them easier to interpret. 

They help in identifying outliers in a regression analysis.

Mathematically, the studentized residual for the $ i $-th observation, denoted as $ t_i $, is calculated as:

$
t_i = \frac{r_i}{s_i}
$

Where:

- $ r_i $ is the residual for the $ i $-th observation, i.e., the difference between the observed and predicted value: $ r_i = y_i - \hat{y}_i $.

- $ s_i $ is the standard error of the residual for the $ i $-th observation, which accounts for the variability in the residuals.

The studentized residuals are useful because they normalize the residuals, allowing us to compare them across different observations. 

If the absolute value of a studentized residual is significantly large, it may indicate that the corresponding observation is an outlier.

### Identifying Outliers Using the Plot of Studentized Residuals vs Index

Now, I'll create a plot of studentized residuals versus the index of observations and use a threshold of 3 to identify potential outliers. 

This threshold is based on the idea that, in a normal distribution, approximately 99.7% of the data should fall within ±3 standard deviations from the mean. 

Hence, any residual with an absolute value greater than 3 is considered unusually large and indicative of a potential outlier.

Let's generate the plot and identify any outliers.


```python
# Calculate studentized residuals
influence = ols_model.get_influence()
studentized_residuals = influence.resid_studentized_internal

# Plot studentized residuals vs index
plt.figure(figsize=(10, 6))
plt.plot(studentized_residuals, 'o', label='Studentized Residuals')
plt.axhline(y=3, color='red', linestyle='--', label='Threshold (+3)')
plt.axhline(y=-3, color='red', linestyle='--', label='Threshold (-3)')

# Annotate potential outliers
for i, residual in enumerate(studentized_residuals):
    if abs(residual) > 3:
        plt.annotate(i, (i, residual), textcoords="offset points", xytext=(0,10), ha='center')

plt.title('Studentized Residuals vs Index')
plt.xlabel('Index')
plt.ylabel('Studentized Residual')
plt.legend()
plt.show()

# Identify the index of potential outliers
potential_outliers = [i for i, residual in enumerate(studentized_residuals) if abs(residual) > 3]
potential_outliers
```


### Interpretation of the Plot

The plot displays the studentized residuals for each observation in the dataset against their index. 

The red dashed lines at +3 and -3 represent the threshold used to identify potential outliers. According to the plot:

- **Observation 27** has a studentized residual greater than 3, indicating that it is a potential outlier.

### Where Does the Threshold Come From?

The threshold of 3 is based on the properties of a normal distribution, where approximately 99.7% of the data should fall within ±3 standard deviations from the mean. 

Using this threshold, any residual with an absolute value greater than 3 is considered unusually large and may suggest that the corresponding observation is an outlier.

### Should Regression Diagnostics Be Repeated After Removing Potential Outliers?

Yes, regression diagnostics should be repeated after removing potential outliers. 

Removing outliers can significantly change the regression model, including the estimated coefficients and the overall fit. 

By repeating the diagnostics, you can ensure that the model is still valid and that no new issues have arisen after the outliers have been removed. 

This step helps to confirm that the final model provides an accurate and reliable representation of the data.

## **High-Leverage Points**

### Understanding High-Leverage Points

**High-leverage points** are observations in your dataset that have an unusual combination of predictor values, meaning they are far from the center of the data's distribution in the space of the independent variables. 

These points can have a disproportionately large influence on the estimation of the regression coefficients because they "pull" the regression line towards themselves.

Mathematically, the leverage $ h_i $ of an observation is derived from the hat matrix $ \mathbf{H} $ in regression analysis:

$
h_i = \mathbf{x}_i^T (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{x}_i
$

Where:

- $ \mathbf{x}_i $ is the vector of the $ i $-th observation's independent variables.

- $ \mathbf{X} $ is the matrix of all independent variables.

The leverage value $ h_i $ ranges from $ \frac{1}{n} $ (where $ n $ is the number of observations) to 1. 

A common rule of thumb is that an observation is considered a high-leverage point if its leverage exceeds $ \frac{2p}{n} $, where $ p $ is the number of predictors (including the intercept) in the model.

### Understanding Cook's Distance

**Cook's Distance** is a measure used to identify influential data points that have a significant impact on the estimated regression coefficients. 

Cook's Distance combines information on both the residuals (how far the predicted values are from the actual values) and the leverage (how far the predictor values are from the center of the data).

Mathematically, Cook's Distance for the $ i $-th observation is defined as:

$
D_i = \frac{1}{p} \cdot \frac{r_i^2}{\hat{\sigma}^2} \cdot \frac{h_i}{(1 - h_i)^2}
$

Where:

- $ r_i $ is the residual for the $ i $-th observation.

- $ \hat{\sigma}^2 $ is the estimated variance of the residuals.

- $ h_i $ is the leverage of the $ i $-th observation.

- $ p $ is the number of predictors (including the intercept).

**The basic idea of Cook's Distance** is to measure how much the regression coefficients would change if a particular observation were removed from the dataset. 

A large Cook's Distance indicates that the observation has a significant influence on the model. 

A commonly used threshold for identifying influential points is $ \frac{4}{n} $, where $ n $ is the number of observations.

### Identifying High-Leverage Points and Influential Points Using Cook's Distance

## **Non-Linearity**

### Understanding Non-Linearity

**Non-linearity** refers to a situation where the relationship between the independent variables (predictors) and the dependent variable (response) is not linear. 

In a linear regression model, we assume that the relationship between the predictors and the response is linear, meaning that changes in the predictors are associated with proportional changes in the response. 

However, if the relationship is non-linear, this assumption is violated, and the model may not accurately represent the data, leading to biased or misleading results.

In a non-linear relationship, the effect of the predictors on the response variable may change at different levels of the predictors, and this change is not constant. 

This can cause problems in regression analysis because the linear model is not flexible enough to capture these complex relationships.

### Understanding the Rainbow Test

The **Rainbow Test** is a statistical test used to detect non-linearity in the relationship between the predictors and the response variable in a regression model. 

The basic idea is to split the data into different parts and compare the fit of the model in these parts to detect any systematic deviations from linearity.

The test works as follows:

- The data is divided into two or more subsets (often, the first and last halves of the sorted data).

- The regression model is fitted to each subset, and the fit is compared.

- If the fit differs significantly between subsets, it suggests that the relationship between the variables may not be linear.

Mathematically, the Rainbow Test involves the following steps:

1. **Fit the Full Model**: Fit the linear regression model to the entire dataset.

2. **Divide the Data**: Split the data into different subsets (often based on the predictor values).

3. **Fit Subset Models**: Fit the model to each subset separately.

4. **Compare Models**: Use an F-test to compare the variance explained by the full model with the variance explained by the subset models.

The test statistic is:

$
F = \frac{\text{RSS}_{\text{full}} - \text{RSS}_{\text{subsets}}}{\text{RSS}_{\text{subsets}} / (n - 2p)}
$

Where:

- $ \text{RSS}_{\text{full}} $ is the residual sum of squares of the full model.

- $ \text{RSS}_{\text{subsets}} $ is the residual sum of squares of the subset models.

- $ n $ is the number of observations.

- $ p $ is the number of predictors.

## **Heteroscedasticity**

### Understanding Heteroscedasticity

**Heteroscedasticity** refers to a condition in regression analysis where the variance of the residuals (errors) is not constant across all levels of the independent variables. 

In a well-fitting linear regression model, one of the key assumptions is **homoscedasticity**, which means that the residuals should have constant variance (i.e., the spread of residuals should be roughly the same across all levels of the fitted values).

When heteroscedasticity is present, the variance of the residuals changes, often increasing or decreasing as the fitted values change. 

This can lead to inefficiency in the estimation of the regression coefficients and can result in biased statistical tests, making it harder to make valid inferences.

### Understanding the Breusch-Pagan Test

The **Breusch-Pagan Test** is a statistical test used to detect heteroscedasticity in a regression model. 

It is specifically designed to test whether the variance of the residuals depends on the values of the independent variables.

The basic idea of the Breusch-Pagan Test is as follows:

1. **Fit the Original Model:** Fit the regression model and obtain the residuals.

2. **Auxiliary Regression:** Regress the squared residuals from the original model on the independent variables.

3. **Test Statistic:** Calculate a test statistic from this auxiliary regression to see if the independent variables explain the variance of the residuals. If they do, it indicates heteroscedasticity.

Mathematically, the test proceeds as follows:

1. **Fit the Regression Model:** Obtain the residuals $ \hat{\epsilon}_i $ from the original model.

2. **Calculate the Squared Residuals:** Compute $ \hat{\epsilon}_i^2 $ for each observation.

3. **Auxiliary Regression:** Regress $ \hat{\epsilon}_i^2 $ on the independent variables $ x_{ij} $:

   $
   \hat{\epsilon}_i^2 = \gamma_0 + \gamma_1 x_{i1} + \gamma_2 x_{i2} + \dots + \gamma_p x_{ip} + v_i
   $

4. **Test Statistic:** The test statistic is calculated as:

   $
   \text{BP} = \frac{n \cdot R^2}{2}
   $

   Where:

   - $ n $ is the number of observations.

   - $ R^2 $ is the coefficient of determination from the auxiliary regression.

5. **Decision:** The test statistic follows a chi-square distribution with degrees of freedom equal to the number of independent variables. 

If the p-value is below a chosen significance level (e.g., 0.05), we reject the null hypothesis and conclude that heteroscedasticity is present.

### Running the Breusch-Pagan Test and Scale-Location Plot

## **Correlation of Error Terms**

### Understanding Correlation of Error Terms

**Correlation of error terms** refers to a situation in regression analysis where the residuals (errors) from one observation are correlated with the residuals from another observation. 

This violates one of the key assumptions of linear regression: that the residuals should be independent of each other.

When the error terms are correlated, it typically indicates the presence of **autocorrelation**. 

Autocorrelation is common in time series data, where the value at one time point is likely to be related to the value at the previous time point. 

If this assumption is violated, it can lead to inefficient estimates of the regression coefficients and biased statistical tests.

### Understanding the Durbin-Watson Test

The **Durbin-Watson Test** is a statistical test used to detect the presence of autocorrelation in the residuals from a regression analysis. 

Specifically, it tests for the presence of first-order autocorrelation, where the residuals are correlated with the immediately preceding residual.

#### Basic Idea

The Durbin-Watson test statistic, $ d $, is calculated as:

$
d = \frac{\sum_{i=2}^{n} (\epsilon_i - \epsilon_{i-1})^2}{\sum_{i=1}^{n} \epsilon_i^2}
$

Where:

- $ \epsilon_i $ is the residual at time $ i $.

- $ n $ is the number of observations.

The value of $ d $ ranges between 0 and 4:

- $ d = 2 $ indicates no autocorrelation.

- $ d < 2 $ indicates positive autocorrelation.

- $ d > 2 $ indicates negative autocorrelation.

#### Decision Rule

- A value of $ d $ between **1.5 and 2.5** is typically considered to indicate that autocorrelation is not a concern.

- Values outside this range suggest potential problems with autocorrelation in the residuals.

## **Normality of Residuals**

### Understanding Normality of Residuals

**Normality of residuals** is an assumption in linear regression that the residuals (errors) follow a normal distribution. 

This assumption is important because it underpins the validity of many statistical tests used to evaluate the significance of the regression coefficients, as well as the overall model fit. 

When residuals are normally distributed, it means that the predictions from the model are accurate on average, with errors being symmetrically distributed around zero.

If the residuals are not normally distributed, it can lead to biased estimates of the regression coefficients, incorrect confidence intervals, and invalid hypothesis tests.

### Understanding the Shapiro-Wilk Test

The **Shapiro-Wilk Test** is a statistical test that checks whether a sample of data comes from a normally distributed population. 

It is particularly powerful for small sample sizes.

#### Basic Idea

The Shapiro-Wilk test calculates a test statistic $ W $, which measures how well the data follow a normal distribution. 

The test compares the order statistics (sorted data points) to the expected values under a normal distribution.

Mathematically, the Shapiro-Wilk test statistic $ W $ is calculated as:

$
W = \frac{\left(\sum_{i=1}^{n} a_i x_{(i)}\right)^2}{\sum_{i=1}^{n} \left(x_i - \bar{x}\right)^2}
$

Where:

- $ x_{(i)} $ is the $ i $-th ordered value (from smallest to largest) in the sample.

- $ \bar{x} $ is the mean of the sample.

- $ a_i $ are constants that depend on the sample size and are derived from the expected values of the order statistics of a standard normal distribution.

#### Decision Rule

- The test statistic $ W $ ranges from 0 to 1, with 1 indicating that the data perfectly follows a normal distribution.

- The null hypothesis $ H_0 $ is that the data is normally distributed.

- If the p-value is less than a chosen significance level (e.g., 0.05), we reject the null hypothesis, indicating that the data is not normally distributed.

### Running the Shapiro-Wilk Test and QQ Plot of Standardized Residuals

Now, I’ll perform the Shapiro-Wilk test to check for normality of the residuals and create a QQ plot to visually inspect the normality of the standardized residuals.


```python
from scipy import stats

# Perform the Shapiro-Wilk test
shapiro_statistic, shapiro_p_value = stats.shapiro(standardized_residuals)

# Generate the QQ plot for standardized residuals
plt.figure(figsize=(10, 6))
stats.probplot(standardized_residuals, dist="norm", plot=plt)
plt.title('QQ Plot of Standardized Residuals')
plt.show()

shapiro_statistic, shapiro_p_value
```

## **Collinearity of Predictors**

### Understanding Collinearity of Predictors

**Collinearity of predictors** refers to a situation in regression analysis where two or more independent variables (predictors) are highly correlated with each other. 

This means that one predictor can be linearly predicted from the others with a substantial degree of accuracy. 

Collinearity can create problems in a regression model because it makes it difficult to isolate the individual effect of each predictor on the dependent variable. 

This can lead to:

- **Unstable Coefficients:** Small changes in the data can cause large changes in the estimated coefficients.

- **Inflated Standard Errors:** The precision of the estimated coefficients decreases, making it harder to determine which predictors are significant.

### Understanding the Variance Inflation Factor (VIF)

The **Variance Inflation Factor (VIF)** is a measure used to quantify the severity of multicollinearity in a regression model. 

VIF assesses how much the variance of a regression coefficient is inflated due to collinearity with other predictors.

#### Basic Idea

VIF is calculated for each predictor by fitting a regression model where the predictor in question is treated as the dependent variable, and all other predictors are treated as independent variables.

 The VIF for each predictor $ j $ is given by:

$
\text{VIF}_j = \frac{1}{1 - R_j^2}
$

Where:

- $ R_j^2 $ is the coefficient of determination from the regression of the $ j $-th predictor on all the other predictors.

#### Interpretation

- **VIF = 1:** No correlation between the $ j $-th predictor and the other predictors.

- **1 < VIF < 5:** Moderate correlation that generally does not warrant concern.

- **VIF > 5:** High correlation that might be problematic.

- **VIF > 10:** Indicates significant multicollinearity, and the predictor is highly correlated with others, potentially leading to unreliable estimates.

## **Overall Information on Statistical Tests**

### Statistical Tests

A **statistical test** is a method used to make decisions or inferences about a population based on sample data. 

It compares the observed data against a null hypothesis (usually a statement of no effect or no difference) to determine if there's enough evidence to reject the null hypothesis in favor of an alternative hypothesis.

### P-Values

A **p-value** is the probability of obtaining the observed data, or something more extreme, assuming that the null hypothesis is true. It's used to determine the significance of the results:

- **Small p-value (typically < 0.05):** Indicates strong evidence against the null hypothesis, leading to its rejection.

- **Large p-value (typically > 0.05):** Indicates weak evidence against the null hypothesis, so it is not rejected.

### Sample Size Considerations

**Sample size** plays a crucial role in the power of a statistical test:

- **Larger Sample Size:** Increases the power of the test, making it more likely to detect a true effect if one exists.

- **Smaller Sample Size:** Reduces the power of the test, making it harder to detect a true effect and increasing the risk of a Type II error (failing to reject a false null hypothesis).

**Power** is the probability that a test will correctly reject a false null hypothesis (i.e., detect an effect when there is one). Ensuring an adequate sample size is essential for obtaining reliable and valid results from statistical tests.
