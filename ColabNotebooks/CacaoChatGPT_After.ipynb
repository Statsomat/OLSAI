# **Part 1: Explanatory Data Analysis**

Descriptive statistics are a set of brief descriptive coefficients that summarize a given data set. 

This summary can be either a representation of the entire population or a sample of it. 

Descriptive statistics are broken down into measures of central tendency and measures of variability (spread). 

### Measures of Central Tendency

These measures describe the center of a data set:

- **Mean**: The average of all the data points.

- **Median**: The middle value when the data points are ordered from smallest to largest.

- **Mode**: The most frequently occurring value in the data set.

### Measures of Variability (Spread)

These measures describe the spread or dispersion within a data set:

- **Range**: The difference between the highest and lowest values.

- **Variance**: A measure of how much the data points differ from the mean.

- **Standard Deviation**: The square root of the variance, providing a measure of the average distance of each data point from the mean.

- **Interquartile Range (IQR)**: The range between the first quartile (25th percentile) and the third quartile (75th percentile).

### Additional Descriptive Statistics

- **Count**: The number of observations in the data set.

- **Minimum and Maximum**: The smallest and largest values in the data set.

- **Percentiles**: Values that divide the data set into equal parts, often used to understand the distribution of the data.

Descriptive statistics provide a simple summary about the sample and the measures. 

Together with simple graphical analysis, they form the basis of virtually every quantitative analysis of data.

### What are Histograms?

Histograms are graphical representations of the distribution of numerical data. 

They are used to understand the shape, central tendency, and variability of a dataset. 

A histogram divides the data into bins (or intervals) and counts the number of data points that fall into each bin.

### Components of a Histogram

1. **Bins (or Intervals)**: The range of values is divided into intervals, called bins. Each bin covers a specific range of data values. The bins should be contiguous, meaning there should be no gaps between them.

2. **Frequency**: The number of data points that fall into each bin. This is represented by the height of the bars in the histogram.

3. **Bars**: Rectangles that represent the frequency of data points in each bin. The height of the bar corresponds to the number of data points in that bin.

4. **X-Axis (Horizontal Axis)**: Represents the bins, or the range of data values divided into intervals.

5. **Y-Axis (Vertical Axis)**: Represents the frequency of data points within each bin.

### How to Interpret a Histogram

1. **Shape**: Look at the overall shape of the histogram.

   - **Symmetrical**: If the left and right sides of the histogram are approximately mirror images of each other, the data is symmetrically distributed.

   - **Skewed Right (Positively Skewed)**: If the histogram has a long tail on the right side, the data is skewed to the right.

   - **Skewed Left (Negatively Skewed)**: If the histogram has a long tail on the left side, the data is skewed to the left.

   - **Uniform**: If all the bars are approximately the same height, the data is uniformly distributed.

   - **Bimodal or Multimodal**: If there are two or more peaks (modes), the data has multiple modes.

2. **Central Tendency**: Identify where the center of the data lies, which can be indicated by the highest bars or the peak of the histogram.

3. **Variability**: Look at the spread of the histogram to understand the variability of the data. Wider histograms indicate more variability, while narrower histograms indicate less variability.

4. **Outliers**: Notice any bars that stand apart from the rest of the histogram. These may represent outliers or unusual data points.

5. **Frequency Distribution**: Assess how often values occur within the dataset by looking at the height of the bars. Taller bars indicate more frequent values within that bin.

Histograms are powerful tools for visualizing and understanding the distribution and spread of data in a dataset. 

They help in identifying patterns, trends, and potential outliers that might require further investigation.

### What are Boxplots?

Boxplots, also known as box-and-whisker plots, are graphical representations that show the distribution of a dataset based on a five-number summary. 

They provide a visual summary of the minimum, first quartile, median, third quartile, and maximum of the data. 

Boxplots are particularly useful for identifying outliers and understanding the spread and skewness of the data.

### Components of a Boxplot

1. **Box**: The main part of the plot, representing the interquartile range (IQR), which contains the middle 50% of the data.

   - **First Quartile (Q1)**: The 25th percentile of the data.

   - **Third Quartile (Q3)**: The 75th percentile of the data.

2. **Median Line**: A line inside the box that represents the median (50th percentile) of the data.

3. **Whiskers**: Lines that extend from the box to the smallest and largest values within 1.5 times the IQR from the quartiles.

   - **Lower Whisker**: Extends from Q1 to the smallest value within 1.5 times the IQR below Q1.

   - **Upper Whisker**: Extends from Q3 to the largest value within 1.5 times the IQR above Q3.

4. **Outliers**: Data points that fall outside the whiskers. These are often plotted as individual points beyond the whiskers.

5. **Minimum**: The smallest value within the whisker range (not considering outliers).
   
6. **Maximum**: The largest value within the whisker range (not considering outliers).

### How to Interpret a Boxplot

1. **Center**: The median line within the box indicates the central tendency of the data. 

The position of the median line relative to the box can give insights into the skewness of the data.

   - If the median is centered, the data is symmetric.

   - If the median is closer to the lower quartile, the data is positively skewed (right-skewed).
   
   - If the median is closer to the upper quartile, the data is negatively skewed (left-skewed).

2. **Spread**: The length of the box (IQR) shows the spread of the middle 50% of the data. A larger box indicates greater variability, while a smaller box indicates less variability.

3. **Whiskers**: The length of the whiskers provides information about the spread of the data outside the middle 50%. Short whiskers indicate that the data outside the quartiles is not very spread out, while long whiskers indicate greater spread.

4. **Outliers**: Points beyond the whiskers are considered outliers. These can indicate variability, measurement errors, or unique properties of the dataset.

5. **Range**: The distance between the minimum and maximum values (excluding outliers) shows the overall range of the data.

Boxplots are valuable for comparing distributions between several groups or datasets, as they provide a clear summary of the data's central tendency, spread, and outliers in a compact form.

### What are ECDF Plots?

ECDF stands for Empirical Cumulative Distribution Function. 

An ECDF plot is a graphical representation that shows the proportion or count of observations falling below each unique value in a dataset. 

Unlike histograms or boxplots, ECDF plots provide a complete view of the data distribution.

### Components of an ECDF Plot

1. **X-Axis (Horizontal Axis)**: Represents the data values sorted in ascending order.

2. **Y-Axis (Vertical Axis)**: Represents the cumulative proportion or fraction of data points less than or equal to each value on the x-axis. It ranges from 0 to 1.

3. **Data Points**: Each point on the ECDF plot corresponds to an observed data value and its cumulative proportion. The points are plotted at the value of the data on the x-axis and the proportion of observations less than or equal to that value on the y-axis.

### How to Interpret an ECDF Plot

1. **Cumulative Proportion**: For any given value on the x-axis, the corresponding value on the y-axis indicates the proportion of data points that are less than or equal to that value. 

   For example, a y-axis value of 0.6 at an x-axis value of 50 means that 60% of the data points are less than or equal to 50.

2. **Distribution Shape**: The overall shape of the ECDF plot provides insight into the distribution of the data.

   - **Steep Slope**: Indicates a high concentration of data points within a narrow range of values, suggesting low variability.

   - **Gradual Slope**: Indicates a wider spread of data points across a range of values, suggesting higher variability.

   - **Flat Segments**: Horizontal stretches indicate ranges where there are no data points.

3. **Median and Percentiles**: The ECDF plot allows for easy determination of median and other percentiles.

   - **Median**: The value at which the y-axis is 0.5.

   - **Percentiles**: Values corresponding to y-axis values like 0.25 (25th percentile), 0.75 (75th percentile), etc.

4. **Comparisons**: When comparing multiple ECDF plots on the same graph, you can easily see differences in distributions between datasets. Shifts to the left or right indicate differences in central tendency, while differences in slope indicate differences in variability.

5. **Outliers and Extremes**: Data points that appear as sudden jumps at the ends of the ECDF plot indicate potential outliers or extreme values in the dataset.

ECDF plots are powerful tools for understanding the distribution and spread of data and for comparing different datasets. 

They provide a complete and straightforward way to visualize cumulative distributions.

```python
import numpy as np

# Function to plot ECDF
def plot_ecdf(data, column, ax):
    x = np.sort(data[column])
    y = np.arange(1, len(x) + 1) / len(x)
    ax.plot(x, y, marker='.', linestyle='none')
    ax.set_title(column)
    ax.set_xlabel(column)
    ax.set_ylabel('ECDF')

# Create subplots
fig, axes = plt.subplots(nrows=5, ncols=3, figsize=(10, 10))
axes = axes.flatten()

# Plot ECDF for each column
for i, column in enumerate(data.columns):
    plot_ecdf(data, column, axes[i])

# Adjust layout
plt.tight_layout()
plt.show()
```

Here are the ECDF plots for each variable in the dataset. 

These plots provide a visual representation of the cumulative distribution for each variable, helping you understand the proportion of data points below each value.

### What are QQ Plots?

QQ (Quantile-Quantile) plots are graphical tools to assess if a dataset follows a particular distribution, typically a normal distribution. 

The plot compares the quantiles of the data against the quantiles of a specified theoretical distribution. 

If the points on the plot fall approximately along a straight line, it suggests that the data follows the specified distribution.

### Components of a QQ Plot

1. **X-Axis (Theoretical Quantiles)**: Represents the quantiles from the theoretical distribution against which the data is being compared. For example, in a normal QQ plot, these are the quantiles from a standard normal distribution.

2. **Y-Axis (Sample Quantiles)**: Represents the quantiles from the observed data.

3. **Data Points**: Each point on the QQ plot corresponds to a quantile pair, one from the sample data and one from the theoretical distribution.

4. **45-Degree Reference Line**: A diagonal line where the quantiles of the sample distribution equal the quantiles of the theoretical distribution. In a normal QQ plot, this line represents where the data points would lie if the data were perfectly normally distributed.

### How to Interpret a QQ Plot

1. **Straight Line Fit**: If the data points fall approximately along the 45-degree reference line, it indicates that the data follows the theoretical distribution. For a normal QQ plot, this would suggest normality.

   - **Perfect Fit**: All points lie exactly on the line.

   - **Good Fit**: Most points lie close to the line with only minor deviations.

2. **Deviations from the Line**:

   - **Tails Deviations**: Points deviating from the line at the ends indicate departures in the tails of the distribution. For example, heavy tails or light tails compared to the normal distribution.

   - **S-Curve Pattern**: If the points form an S-shape, this indicates that the data has heavier tails than the normal distribution (leptokurtic distribution).

   - **Inverted S-Curve**: If the points form an inverted S-shape, this suggests that the data has lighter tails than the normal distribution (platykurtic distribution).

3. **Outliers**: Points far from the reference line may indicate outliers in the data.

4. **Symmetry and Distribution Shape**: Patterns and deviations from the reference line can provide insights into the skewness and kurtosis of the data distribution.

QQ plots are valuable tools for visually assessing the distribution of data and for identifying departures from expected theoretical distributions. 

They are particularly useful in statistical analysis and hypothesis testing.

```python
import scipy.stats as stats

# Function to plot QQ plot
def plot_qq(data, column, ax):
    stats.probplot(data[column], dist="norm", plot=ax)
    ax.set_title(column)

# Create subplots
fig, axes = plt.subplots(nrows=5, ncols=3, figsize=(10, 10))
axes = axes.flatten()

# Plot QQ plot for each column
for i, column in enumerate(data.columns):
    plot_qq(data, column, axes[i])

# Adjust layout
plt.tight_layout()
plt.show()
```

# **Part 2: Preparations for Regression Diagnostics**

Sure, let's go through the assumptions of a linear regression model in detail.

### 1. Linearity

**Assumption**: The relationship between the independent variables (predictors) and the dependent variable (response) is linear.

**Mathematical Form**: The model is expressed as:

$ y_i = \beta_0 + \beta_1 x_{i1} + \beta_2 x_{i2} + \cdots + \beta_p x_{ip} + \epsilon_i $

where:

- $y_i$ is the dependent variable for observation $i$.

- $\beta_0$ is the intercept.

- $\beta_1, \beta_2, \ldots, \beta_p$ are the coefficients for the independent variables.

- $x_{i1}, x_{i2}, \ldots, x_{ip}$ are the values of the independent variables for observation $ i $.

- $\epsilon_i$ is the error term for observation $i$.

**Technical Terms**:

- **Dependent Variable**: The outcome variable that the model aims to predict.

- **Independent Variables**: The input variables used to predict the dependent variable.

- **Intercept**: The expected value of the dependent variable when all independent variables are zero.

- **Coefficients**: Parameters that measure the change in the dependent variable for a one-unit change in the independent variable, holding all other variables constant.

- **Error Term**: The difference between the observed and predicted values of the dependent variable.

### 2. Independence

**Assumption**: The observations (data points) are independent of each other. This means the value of the error term for any observation is not correlated with the value of the error term for any other observation.

**Mathematical Form**: 

$\text{Cov}(\epsilon_i, \epsilon_j) = 0 \quad \text{for all} \quad i \neq j$

where:

- $\text{Cov}$ denotes covariance.

- $\epsilon_i$ and $\epsilon_j$ are the error terms for observations $i$ and $j$, respectively.

**Technical Terms**:

- **Covariance**: A measure of the degree to which two variables change together. If the covariance is zero, the variables are independent.

### 3. Homoscedasticity

**Assumption**: The variance of the error terms is constant across all levels of the independent variables.

**Mathematical Form**:

$\text{Var}(\epsilon_i) = \sigma^2 \quad \text{for all} \quad i$

where:

- $\text{Var}$ denotes variance.

- $\sigma^2$ is a constant.

**Technical Terms**:

- **Variance**: A measure of the dispersion of a set of values. It is the average of the squared differences from the mean.

- **Homoscedasticity**: The condition of having constant variance in the error terms.

### 4. No Perfect Multicollinearity

**Assumption**: No independent variable is a perfect linear function of any other independent variable(s).

**Mathematical Form**: For any independent variables $x_{ij}$ and $x_{ik}$,

$\text{if} \quad x_{ij} = \alpha_0 + \alpha_1 x_{ik} \quad \text{then} \quad \alpha_1 = 0$

where:

- $\alpha_0$ and $\alpha_1$ are constants.

**Technical Terms**:

- **Multicollinearity**: A situation where independent variables are highly linearly related, making it difficult to isolate the individual effect of each variable on the dependent variable.

### 5. Normality of Errors

**Assumption**: The error terms are normally distributed. This is particularly important for hypothesis testing and constructing confidence intervals.

**Mathematical Form**:

$\epsilon_i \sim \mathcal{N}(0, \sigma^2)$

where:

- $\mathcal{N}(0, \sigma^2)$ denotes a normal distribution with mean 0 and variance $\sigma^2$.

**Technical Terms**:

- **Normal Distribution**: A symmetric, bell-shaped distribution characterized by its mean and variance.

- **Mean**: The average of all values in a dataset.

- **Standard Deviation**: The square root of the variance, indicating the dispersion of a dataset.

### Summary of Equations

1. **Linear Relationship**:
   $y_i = \beta_0 + \beta_1 x_{i1} + \beta_2 x_{i2} + \cdots + \beta_p x_{ip} + \epsilon_i$

2. **Independence**:
   $\text{Cov}(\epsilon_i, \epsilon_j) = 0 \quad \text{for all} \quad i \neq j$

3. **Homoscedasticity**:
   $\text{Var}(\epsilon_i) = \sigma^2 \quad \text{for all} \quad i$

4. **No Perfect Multicollinearity**:
   $\text{if} \quad x_{ij} = \alpha_0 + \alpha_1 x_{ik} \quad \text{then} \quad \alpha_1 = 0$

5. **Normality of Errors**:
   $\epsilon_i \sim \mathcal{N}(0, \sigma^2)$

Understanding these assumptions is crucial for ensuring that the results from a linear regression model are valid and reliable. If any of these assumptions are violated, the model's estimates, predictions, and inferences may be incorrect.

# **Part 3: Regression Diagnostics**

Regression diagnostics involve a set of procedures used to evaluate the validity and performance of a regression model. 

The primary goal of regression diagnostics is to ensure that the assumptions underlying the regression model are met and to identify any potential issues that might affect the accuracy and reliability of the model's estimates and predictions.

### Key Objectives of Regression Diagnostics

1. **Assessing Model Assumptions**: Regression diagnostics help in verifying whether the assumptions of the regression model (such as linearity, independence, homoscedasticity, and normality) hold true for the given data. Meeting these assumptions is crucial for the validity of the model.

2. **Identifying Model Fit**: Diagnostics evaluate how well the regression model fits the data. This includes checking if the model accurately captures the relationship between the dependent and independent variables.

3. **Detecting Outliers and Influential Points**: Regression diagnostics aim to identify data points that are outliers or have a disproportionately large influence on the model's parameters. These points can distort the results and lead to incorrect conclusions.

4. **Evaluating Residuals**: The residuals (differences between observed and predicted values) are examined to ensure they exhibit random patterns without systematic structures. Patterns in residuals can indicate model inadequacies or violations of assumptions.

5. **Checking for Multicollinearity**: Diagnostics help in identifying if there is a high correlation among independent variables, which can make it difficult to isolate the effect of each variable and lead to unreliable estimates.

6. **Improving Model Specification**: By identifying issues such as omitted variables or incorrect functional forms, regression diagnostics guide the improvement of model specification to enhance its explanatory power and predictive accuracy.

### Importance of Regression Diagnostics

- **Validity of Inferences**: Proper diagnostics ensure that the statistical inferences drawn from the model (such as hypothesis tests and confidence intervals) are valid and reliable.

- **Model Robustness**: Diagnostics help in building robust models that perform well not only on the given data but also on new, unseen data.

- **Decision-Making**: Reliable regression models are critical for making informed decisions based on the relationships identified between variables.

- **Efficiency of Estimations**: Ensuring that the assumptions are met and identifying potential issues lead to more efficient and unbiased parameter estimates.

In summary, regression diagnostics are essential tools for evaluating the performance and validity of a regression model, ensuring the results are trustworthy and useful for analysis and decision-making.

## **Outliers**

**Outliers** are data points that differ significantly from other observations in a dataset. 

They can arise due to variability in the data, measurement errors, or unusual occurrences. 

Outliers can have a substantial impact on the results of statistical analyses, including regression models, as they can skew the results and affect the model's accuracy and assumptions.

### Studentized Residuals

**Studentized residuals** are a form of residuals used in regression analysis, adjusted for the variability of each data point. They are used to identify outliers and assess the fit of a regression model.

**Mathematical Equations**:
1. **Residual ($r_i$)**: The difference between the observed and predicted values.
   $r_i = y_i - \hat{y}_i$
   where $y_i$ is the observed value and $\hat{y}_i$ is the predicted value for observation $i$.

2. **Standardized Residual ($e_i$)**: The residual divided by an estimate of its standard deviation.
   $e_i = \frac{r_i}{\hat{\sigma} \sqrt{1 - h_i}}$
   where $\hat{\sigma}$ is the estimated standard deviation of the residuals, and $h_i$ is the leverage of observation $i$.

3. **Studentized Residual ($t_i$)**: A further adjusted form of standardized residual, taking into account the variance inflation.
   $t_i = \frac{r_i}{\hat{\sigma}_i \sqrt{1 - h_i}}$
   where $\hat{\sigma}_i$ is the estimated standard deviation of the residuals excluding observation $i$.

### Plotting and Interpreting Studentized Residuals

## **High-Leverage Points**

**High-leverage points** are observations in a dataset that have extreme values for one or more independent variables. 

These points can significantly influence the fit of a regression model because they can disproportionately affect the estimation of the regression coefficients.

**Leverage**: The leverage of an observation measures how far away the values of the independent variables are from the mean values of the independent variables. 

It is derived from the hat matrix $\mathbf{H}$, which is calculated as:

$\mathbf{H} = \mathbf{X}(\mathbf{X}^T\mathbf{X})^{-1}\mathbf{X}^T$

where $\mathbf{X}$ is the matrix of independent variables.

The diagonal elements $h_i$ of the hat matrix $\mathbf{H}$ represent the leverage of each observation:

$h_i = \mathbf{x}_i^T(\mathbf{X}^T\mathbf{X})^{-1}\mathbf{x}_i$

### Cook's Distance

**Cook's Distance** is a measure used in regression analysis to identify influential data points that have a significant impact on the estimated regression coefficients. 

It combines the information from residuals and leverage to determine the influence of each observation.

**Mathematical Equation**:

$D_i = \frac{r_i^2}{p \hat{\sigma}^2} \cdot \frac{h_i}{(1 - h_i)^2}$

where:

- $r_i$ is the residual for observation $i$.

- $ p$ is the number of parameters in the model (including the intercept).

- $ \hat{\sigma}^2 $ is the mean squared error of the regression model.

- $ h_i $ is the leverage of observation $ i $.

### Plotting and Interpreting Cook's Distance

Let's plot Cook's Distance versus the index of observations and identify high-leverage points using a threshold of $\frac{4}{n}$, where $n $ is the number of observations.

#### Python Code to Create the Plot and Identify High-Leverage Points

## **Non-Linearity**

**Non-linearity** in the context of regression analysis refers to a situation where the relationship between the independent variables (predictors) and the dependent variable (response) is not linear. 

In a linear regression model, it is assumed that the change in the dependent variable is a linear function of the changes in the independent variables. 

When this assumption is violated, the model may not adequately capture the true relationship, leading to biased or misleading results.

### Rainbow Test

The **Rainbow Test** is a statistical test used to detect non-linearity in a regression model. 

It assesses whether the relationship between the dependent and independent variables is indeed linear. 

The test works by splitting the data into two parts, fitting separate models to each part, and then comparing the residuals of these models.

**Mathematical Equations**:

1. **Fit the Model**: Fit the full model to the data:

    $y_i = \beta_0 + \beta_1 x_{i1} + \beta_2 x_{i2} + \cdots + \beta_p x_{ip} + \epsilon_i$
   
2. **Split the Data**: Split the data into two subsets.

3. **Fit Subset Models**: Fit separate models to each subset and compute the residuals.

4. **Compute the Test Statistic**: Compare the residuals from the subset models with the residuals from the full model.

The test statistic is derived from an F-distribution and is used to determine whether the differences between the subset models and the full model are statistically significant.

### Conducting the Rainbow Test and Plotting Residuals vs. Fitted Values

Let's conduct the Rainbow Test and plot the residuals versus the fitted values to assess non-linearity.

#### Python Code to Conduct the Rainbow Test and Create the Plot

## **Heteroscedasticity**

**Heteroscedasticity** refers to a situation in regression analysis where the variance of the residuals (errors) is not constant across all levels of the independent variables. 

In other words, the spread of the residuals changes as the value of an independent variable changes. 

This violates the assumption of homoscedasticity, which assumes that the residuals have constant variance. 

Heteroscedasticity can lead to inefficient estimates and biased standard errors, affecting hypothesis tests and confidence intervals.

### Breusch-Pagan Test

The **Breusch-Pagan Test** is a statistical test used to detect heteroscedasticity in a regression model. It tests whether the variance of the residuals is dependent on the values of the independent variables.

**Mathematical Equations**:

1. **Fit the Original Regression Model**:

   $y_i = \beta_0 + \beta_1 x_{i1} + \beta_2 x_{i2} + \cdots + \beta_p x_{ip} + \epsilon_i$

   Obtain the residuals $\epsilon_i$.

2. **Auxiliary Regression**: Regress the squared residuals $\epsilon_i^2$ on the independent variables:

   $\epsilon_i^2 = \alpha_0 + \alpha_1 x_{i1} + \alpha_2 x_{i2} + \cdots + \alpha_p x_{ip} + u_i$

3. **Test Statistic**: The test statistic $LM$ is given by:

   $LM = nR^2$

   where $n$ is the number of observations and $R^2$ is the coefficient of determination from the auxiliary regression.

4. **Distribution**: The test statistic $LM$ follows a chi-square distribution with $p$ degrees of freedom, where $p$ is the number of independent variables.

### Conducting the Breusch-Pagan Test and Creating the Scale-Location Plot

Let's conduct the Breusch-Pagan Test and plot the scale-location plot to assess heteroscedasticity.

#### Python Code to Conduct the Breusch-Pagan Test and Create the Plot


```python
from statsmodels.stats.diagnostic import het_breuschpagan

# Conduct the Breusch-Pagan Test
bp_test = het_breuschpagan(model.resid, model.model.exog)

# Extract the test statistic and p-value
bp_stat = bp_test[0]
bp_p_value = bp_test[1]

# Plot Scale-Location Plot
fitted_values = model.fittedvalues
standardized_residuals = model.get_influence().resid_studentized_internal

plt.figure(figsize=(10, 6))
plt.scatter(fitted_values, np.sqrt(np.abs(standardized_residuals)), alpha=0.5)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('Fitted Values')
plt.ylabel('Square Root of Standardized Residuals')
plt.title('Scale-Location Plot')
plt.show()

bp_stat, bp_p_value
```

### Interpretation of Heteroscedasticity Test and Scale-Location Plot

**Breusch-Pagan Test Results**:

- **p-value**: 0.0733

The p-value for the Breusch-Pagan Test is 0.0733, which is slightly above the common significance level of 0.05. 

This suggests that there is not strong evidence to reject the null hypothesis of homoscedasticity. Therefore, we do not have strong evidence of heteroscedasticity based on this test.

## **Correlation of Error Terms**

**Correlation of error terms** (also known as autocorrelation) occurs when the residuals (errors) in a regression model are not independent of each other. 

This means that the value of the residual for one observation is correlated with the value of the residual for another observation. 

This violation of the independence assumption can lead to inefficient estimates and affect the validity of statistical tests.

### Durbin-Watson Test

The **Durbin-Watson Test** is a statistical test used to detect the presence of autocorrelation in the residuals of a regression model. 

It specifically tests for first-order autocorrelation, which means it checks whether the residuals are correlated with their immediately preceding values.

**Mathematical Equations**:

1. **Durbin-Watson Statistic (DW)**:

   $DW = \frac{\sum_{i=2}^{n} (e_i - e_{i-1})^2}{\sum_{i=1}^{n} e_i^2}$
   
   where:

   - $e_i$ is the residual for observation $i$.

   - $n$ is the number of observations.

2. **Interpretation of DW Statistic**:

   - $DW \approx 2$: No autocorrelation.

   - $DW < 2$: Positive autocorrelation.

   - $DW > 2 $: Negative autocorrelation.

### Conducting the Durbin-Watson Test and Plotting Residuals Over Time

Let's conduct the Durbin-Watson Test and plot the residuals over time to assess autocorrelation.

#### Python Code to Conduct the Durbin-Watson Test and Create the Plot

```python
from statsmodels.stats.stattools import durbin_watson

# Conduct the Durbin-Watson Test
dw_stat = durbin_watson(model.resid)

# Plot Residuals over Time (Index)
plt.figure(figsize=(10, 6))
plt.plot(model.resid, marker='o', linestyle='-')
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('Index')
plt.ylabel('Residuals')
plt.title('Residuals Over Time')
plt.show()

dw_stat
```

### Interpretation of the Durbin-Watson Test and Residuals Over Time Plot

**Durbin-Watson Test Results**:

- **DW Statistic**: 1.441

The Durbin-Watson statistic is 1.441. Interpreting this value:

- A DW value of approximately 2 suggests no autocorrelation.

- A DW value significantly less than 2 indicates positive autocorrelation.

- A DW value significantly greater than 2 indicates negative autocorrelation.

Since the DW value is 1.441, which is less than 2 but not drastically so, it suggests some degree of positive autocorrelation in the residuals.

### Residuals Over Time Plot

**Key Points in the Plot**:

- **Random Scatter**: If the residuals are randomly scattered around the horizontal line, it suggests no autocorrelation.

### Observations

## **Normality of Residuals**

**Normality of residuals** is an assumption in linear regression that the residuals (errors) are normally distributed. 

This assumption is important for conducting hypothesis tests and constructing confidence intervals for the regression coefficients. 

If the residuals are normally distributed, the inference procedures based on the regression model will be valid.

### Shapiro-Wilk Test

The **Shapiro-Wilk Test** is a statistical test used to assess the normality of a dataset. It tests the null hypothesis that the data was drawn from a normal distribution.

**Mathematical Equations**:

1. **Test Statistic (W)**:

   $W = \frac{\left( \sum_{i=1}^n a_i x_{(i)} \right)^2}{\sum_{i=1}^n (x_i - \bar{x})^2}$

   where:

   - $ x_{(i)} $ are the ordered sample values.

   - $ a_i $ are constants derived from the expected values of the order statistics of a standard normal distribution.

   - $ \bar{x}$ is the sample mean.

   - $ n $ is the number of observations.

2. **Interpretation**:

   - A low p-value (typically < 0.05) indicates that the null hypothesis is rejected, suggesting that the residuals are not normally distributed.

   - A high p-value indicates that the null hypothesis is not rejected, suggesting that the residuals are normally distributed.

### Conducting the Shapiro-Wilk Test and Plotting the QQ Plot

Let's conduct the Shapiro-Wilk Test and plot the QQ plot of standardized residuals to assess normality.

#### Python Code to Conduct the Shapiro-Wilk Test and Create the QQ Plot


```python
from scipy.stats import shapiro
import statsmodels.api as sm

# Conduct the Shapiro-Wilk Test
shapiro_stat, shapiro_p_value = shapiro(model.resid)

# Plot QQ Plot of Standardized Residuals
sm.qqplot(model.resid, line='s')
plt.title('QQ Plot of Standardized Residuals')
plt.show()

shapiro_stat, shapiro_p_value
```

### Interpretation of Normality Test and QQ Plot

**Shapiro-Wilk Test Results**:

- **Test Statistic**: 0.9848

- **p-value**: 0.1990

The p-value for the Shapiro-Wilk Test is 0.1990, which is greater than the common significance level of 0.05. 

This indicates that we fail to reject the null hypothesis that the residuals are normally distributed. Therefore, there is no strong evidence against the normality of the residuals.

### QQ Plot of Standardized Residuals

**Key Points in the Plot**:

- **X-Axis**: Theoretical quantiles from a standard normal distribution.

- **45-Degree Line**: This line represents where the points would lie if the residuals were perfectly normally distributed.

**Interpretation**:

### Summary

## **Collinearity of Predictors**

**Collinearity** (or multicollinearity) of predictors refers to a situation in which two or more independent variables in a regression model are highly linearly related. 

This means that one predictor can be almost perfectly predicted from the others. Collinearity can cause several issues:

- It makes it difficult to determine the individual effect of each predictor on the dependent variable.

- It can inflate the standard errors of the regression coefficients, leading to less reliable statistical tests.

- It can cause numerical instability in the estimation of the regression coefficients.

### Variance Inflation Factor (VIF)

The **Variance Inflation Factor (VIF)** is a measure used to quantify the extent of multicollinearity in a set of multiple regression variables. 

VIF assesses how much the variance of an estimated regression coefficient increases when predictors are correlated.

**Mathematical Equations**:

1. **Calculate VIF for each predictor**:

   - Fit a regression model where the predictor $X_i $ is regressed on all other predictors.

   - Calculate the coefficient of determination $R_i^2$ for this regression.

   - Compute VIF for predictor $X_i$:

     $\text{VIF}_i = \frac{1}{1 - R_i^2}$

2. **Interpretation**:
   - $\text{VIF}_i = 1$: No collinearity.

   - $1 < \text{VIF}_i < 5$: Moderate collinearity.

   - $\text{VIF}_i > 5 $: High collinearity, indicating a problem.

### Conducting the VIF Analysis and Plotting the Correlation Matrix

Let's calculate the VIF for each predictor and plot the correlation matrix to assess collinearity.

#### Python Code to Calculate VIF and Plot the Correlation Matrix
