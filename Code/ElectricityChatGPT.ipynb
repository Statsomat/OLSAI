## **Descriptive Statistics**

Descriptive statistics are tools that help us summarize and understand large amounts of data by providing simple numerical descriptions. 

They focus on describing the main features of a dataset in a straightforward way, without drawing conclusions beyond the data itself. 

Think of it as a way to get a quick overview of what the data looks like.

Here are the key types of descriptive statistics:

1. **Measures of Central Tendency**: These tell us where the center of the data lies.

   - **Mean**: The average of all data points. You add up all the values and divide by the number of values.

   - **Median**: The middle value when the data is sorted in order. If there's an even number of data points, it's the average of the two middle values.

   - **Mode**: The value that appears most frequently in the data.

2. **Measures of Dispersion**: These tell us how spread out the data is.

   - **Range**: The difference between the highest and lowest values.

   - **Variance**: A measure of how much the data points differ from the mean. It tells us how spread out the data is.

   - **Standard Deviation**: The square root of the variance, it gives us an idea of how much data points typically deviate from the mean.

3. **Measures of Shape**: These describe the shape or distribution of the data.

   - **Skewness**: It tells us if the data is symmetrical or if it's skewed to the left or right.
   
   - **Kurtosis**: It measures the "tailedness" of the data distribution. High kurtosis means more data is in the tails (outliers), while low kurtosis means more data is in the center.

Descriptive statistics are the first step in data analysis, giving you a quick summary of what your data looks like before diving deeper into more complex analyses.

## **Histograms**

Let's dive into histograms step by step:

### 1. What are Histograms?
A histogram is a type of bar graph that represents the distribution of a dataset. It shows the frequency (or count) of data points that fall within certain ranges, known as "bins." Each bar in a histogram represents a bin, and the height of the bar indicates the number of data points in that range.

### 2. Components of a Histogram
- **Bins**: These are the intervals into which the entire range of data is divided. The bins are usually of equal size, but not always.
- **Bars**: Each bin is represented by a bar. The height of the bar corresponds to the number of data points (frequency) within that bin.
- **X-axis**: Represents the data range divided into bins.
- **Y-axis**: Represents the frequency (or count) of data points in each bin.
- **KDE Line** (Kernel Density Estimate): A smooth curve that estimates the probability density function of the variable. It's optional but provides a clearer view of the data's distribution.

### 3. How to Interpret a Histogram
- **Shape**: Look at the shape of the histogram to understand the distribution of the data. Common shapes include:
  - **Normal Distribution**: Bell-shaped, symmetrical around the mean.
  - **Skewed Left/Right**: A longer tail on one side indicates skewness.
  - **Bimodal**: Two peaks suggest two prevalent ranges of values.
- **Spread**: The width of the bars and the range of the x-axis give an idea of how spread out the data is.
- **Center**: The area where most of the data points are concentrated indicates the central tendency.
- **Outliers**: Bars that are isolated far from the rest of the histogram can indicate outliers.

## **Boxplots**

Let's explore boxplots in detail:

### 1. What are Boxplots?
A boxplot (or box-and-whisker plot) is a standardized way of displaying the distribution of data based on a five-number summary: the minimum, first quartile (Q1), median, third quartile (Q3), and maximum. Boxplots are particularly useful for identifying outliers and understanding the spread and skewness of the data.

### 2. Components of a Boxplot
- **Box**: The main rectangular part of the plot. It represents the interquartile range (IQR), which is the range between the first quartile (Q1) and the third quartile (Q3).
- **Median Line**: A line inside the box that represents the median (Q2) of the data.
- **Whiskers**: Lines extending from the box to the smallest and largest values within 1.5 times the IQR from the quartiles. Whiskers give an indication of the spread of the majority of the data.
- **Outliers**: Data points that fall outside the range of the whiskers are plotted as individual points and considered outliers.
- **Minimum and Maximum**: The smallest and largest data points within the whiskers, excluding outliers.

### 3. How to Interpret a Boxplot
- **Median**: The line inside the box indicates the median (Q2), showing the central tendency of the data.
- **Spread**: The width of the box (IQR) shows the middle 50% of the data. A wider box indicates more variability.
- **Skewness**: If the median line is closer to the top or bottom of the box, the data might be skewed. For example, if the median is closer to Q1, the data is skewed right.
- **Outliers**: Points outside the whiskers are potential outliers. They may indicate unusual data points that could be of interest or need further investigation.
- **Symmetry and Distribution**: If the whiskers are of equal length and the box is symmetrical, the data is likely symmetrically distributed. If not, the data may be skewed.

## **ECDF Plots**

Let's delve into ECDF (Empirical Cumulative Distribution Function) plots:

### 1. What are ECDF Plots?
An ECDF plot is a graphical representation of the cumulative distribution of a dataset. It shows the proportion or percentage of data points that are less than or equal to a certain value. Unlike histograms or boxplots, which provide a summary of the data distribution, ECDF plots give a more detailed view by displaying every data point.

### 2. Components of an ECDF Plot
- **X-axis**: Represents the data values sorted in ascending order.
- **Y-axis**: Represents the cumulative proportion of data points. It ranges from 0 to 1 (or 0% to 100%).
- **Step Line**: The main feature of the ECDF plot is a step-like curve that increases as you move from left to right along the x-axis. Each step represents a data point, and the height of the step indicates the cumulative proportion of data points at that value.

### 3. How to Interpret an ECDF Plot
- **Shape of the Curve**: 
  - A steep rise in the curve indicates a large number of data points concentrated in a narrow range.
  - A gradual rise indicates that the data points are spread out over a wider range.
- **Flat Sections**: These indicate ranges where there are no data points.
- **Proportion**: The y-value at any point on the curve shows the proportion of data points less than or equal to the corresponding x-value.
- **Comparing Distributions**: ECDF plots are particularly useful when comparing multiple datasets. If one curve is consistently above another, it indicates that the first dataset generally has smaller values.

## **QQ Plots**

Let's explore QQ (Quantile-Quantile) plots:

### 1. What are QQ Plots?
A QQ plot is a graphical tool used to compare the distribution of a dataset to a theoretical distribution, such as the normal distribution. It helps assess whether the data follows a specific distribution by plotting the quantiles of the data against the quantiles of the theoretical distribution. If the data follows the distribution, the points should roughly align along a straight line.

### 2. Components of a QQ Plot
- **X-axis (Theoretical Quantiles)**: Represents the quantiles of the theoretical distribution (e.g., normal distribution).
- **Y-axis (Sample Quantiles)**: Represents the quantiles of the actual data.
- **45-Degree Reference Line**: A diagonal line that represents a perfect match between the data and the theoretical distribution. If the data fits the distribution well, the points will lie close to this line.
- **Data Points**: Each point on the plot represents a quantile of the data plotted against the corresponding quantile of the theoretical distribution.

### 3. How to Interpret a QQ Plot
- **Alignment with the 45-Degree Line**:
  - **Close Alignment**: If the points closely follow the 45-degree line, the data is likely to follow the theoretical distribution (e.g., normal distribution).
  - **Deviations**: 
    - **Upward/Downward Curvature**: A curve away from the line (upwards or downwards) suggests that the data has more or fewer extreme values (fat tails or thin tails) than expected in the theoretical distribution.
    - **S-Shaped Curve**: This indicates skewness. An upward curve at the ends with a downward curve in the middle suggests a right-skewed distribution, while the reverse suggests left-skewness.
- **Outliers**: Points far from the line may indicate outliers or data points that do not conform to the distribution.

## Multiple Linear Regression

### What is Multiple Linear Regression?

Multiple linear regression is a statistical technique used to model the relationship between one dependent variable (denoted as $ y $) and multiple independent variables (denoted as $ x_{ij} $). The goal is to find the linear equation that best predicts the dependent variable using the independent variables.

The general form of a multiple linear regression model is:

$
y_i = \beta_0 + \sum_{j=1}^{k} \beta_j x_{ij} + \epsilon_i
$

Here:
- $ y_i $ is the dependent variable for the $ i $-th observation.
- $ x_{ij} $ represents the $ j $-th independent variable for the $ i $-th observation.
- $ \beta_0 $ is the intercept term.
- $ \beta_j $ (for $ j = 1, 2, \dots, k $) are the coefficients that represent the effect of each independent variable on the dependent variable.
- $ \epsilon_i $ is the error term for the $ i $-th observation, which accounts for the variability in $ y $ that cannot be explained by the linear relationship.

### Explanation of $ i $ and $ j $
- **$ i $**: This index ranges over the number of observations in the dataset, typically from $ 1 $ to $ n $, where $ n $ is the total number of observations.
- **$ j $**: This index ranges over the number of independent variables, typically from $ 1 $ to $ k $, where $ k $ is the total number of independent variables.

### Assumptions of a Classical Linear Regression Model

1. **Linearity**:
   - The relationship between the dependent variable $ y_i $ and the independent variables $ x_{ij} $ is linear.
   - Mathematically, the model is expressed as:
   $
   y_i = \beta_0 + \sum_{j=1}^{k} \beta_j x_{ij} + \epsilon_i
   $

2. **Independence**:
   - The observations (i.e., $ y_i $ and $ \epsilon_i $) are independent of each other. This means the error terms $ \epsilon_i $ are not correlated across observations.
   - Mathematically, for any two different observations $ i $ and $ i' $:
   $
   \text{Cov}(\epsilon_i, \epsilon_{i'}) = 0
   $

3. **Homoscedasticity**:
   - The variance of the error terms $ \epsilon_i $ is constant across all levels of the independent variables.
   - Mathematically:
   $
   \text{Var}(\epsilon_i) = \sigma^2 \quad \text{for all } i
   $

4. **Normality**:
   - The error terms $ \epsilon_i $ are normally distributed.
   - Mathematically:
   $
   \epsilon_i \sim \mathcal{N}(0, \sigma^2)
   $

5. **No Perfect Multicollinearity**:
   - The independent variables $ x_{ij} $ are not perfectly correlated with each other. This means no independent variable is a perfect linear combination of the others.
   - Mathematically, there should be no exact linear relationship among the independent variables:
   $
   \text{Rank}(X) = k+1
   $
   where $ X $ is the matrix of independent variables.

### Summary of Assumptions in Mathematical Form
1. **Linearity**: $ y_i = \beta_0 + \sum_{j=1}^{k} \beta_j x_{ij} + \epsilon_i $
2. **Independence**: $ \text{Cov}(\epsilon_i, \epsilon_{i'}) = 0 $ for $ i \neq i' $
3. **Homoscedasticity**: $ \text{Var}(\epsilon_i) = \sigma^2 $
4. **Normality**: $ \epsilon_i \sim \mathcal{N}(0, \sigma^2) $
5. **No Perfect Multicollinearity**: $ \text{Rank}(X) = k+1 $

These assumptions are crucial for ensuring that the estimates of the regression coefficients $ \beta_j $ are unbiased, efficient, and valid for inference.

## Regression Diagnostics

Regression diagnostics are a set of techniques and analyses used to evaluate the validity and reliability of a regression model. They help ensure that the assumptions underlying the regression analysis are met and identify any potential issues that could affect the accuracy of the model's predictions or the interpretation of the results.

The purpose of regression diagnostics is to assess the fit of the model, detect any anomalies or deviations from the assumptions, and guide potential improvements to the model. This process involves examining various aspects of the model, such as the residuals (the differences between observed and predicted values), to determine whether the model accurately represents the data and whether the relationships between variables are correctly specified.

In summary, regression diagnostics help verify that the model is robust, the assumptions are valid, and the results are reliable, enabling you to trust the conclusions drawn from the regression analysis.

To translate the OLS regression model from R into Python, we'll follow these steps:

1. **Prepare the Data**: Load the dataset and preprocess it as done in the R code.

### Steps:

1. **In R**: The model uses the following formula:
   $
   \text{log}(\text{cost}/\text{fuel}) \sim \text{log}(\text{output}) + \text{log}(\text{labor}/\text{fuel}) + \text{log}(\text{capital}/\text{fuel})
   $
   This model is fitted on the first 145 observations of the `Electricity1955` dataset.

2. **In Python**: We'll use the `statsmodels` library to replicate this. The equivalent steps involve importing the necessary libraries, performing the log transformations, and fitting the OLS regression model.

Let's go ahead and implement this in Python.

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

# Select the first 145 rows
Electricity = data.iloc[:145, :]

# Define the independent variables (X) and dependent variable (y)
X = Electricity[['output', 'labor', 'capital']]
y = np.log(Electricity['cost'] / Electricity['fuel'])

# Log-transform the independent variables
X = np.log(X.div(Electricity['fuel'], axis=0))

# Add a constant to the model (for the intercept)
X = sm.add_constant(X)

# Fit the OLS regression model
model = sm.OLS(y, X).fit()

# Model is built but not displaying the summary or parameters as requested
model

```

## **Outliers**

### Understanding Outliers

**Outliers** are data points that deviate significantly from the other observations in a dataset. They can be unusually high or low values compared to the rest of the data. Outliers are important because they can distort statistical analyses, leading to misleading results or conclusions. Identifying and addressing outliers is crucial in regression analysis to ensure the model's accuracy and reliability.

Outliers can occur due to various reasons, including measurement errors, data entry mistakes, or they might represent a rare event. In the context of regression, outliers can affect the estimated coefficients and the goodness-of-fit, making the model less representative of the underlying relationship between variables.

### Understanding Studentized Residuals

**Studentized residuals** are a way of standardizing residuals in regression analysis. Residuals are the differences between observed values and the values predicted by the regression model. Studentized residuals adjust these residuals by taking into account the variability of the data, allowing us to detect outliers more effectively.

The basic idea is that studentized residuals help identify outliers by showing how far a particular observation's residual is from the expected value in standard deviation units.

#### Mathematical Equations

Let:
- $ y_i $ be the observed value for the $ i $-th observation.
- $ \hat{y}_i $ be the predicted value for the $ i $-th observation.
- $ r_i = y_i - \hat{y}_i $ be the residual for the $ i $-th observation.

The studentized residual for the $ i $-th observation is given by:

$
t_i = \frac{r_i}{\hat{\sigma} \sqrt{1 - h_{ii}}}
$

Where:
- $ \hat{\sigma} $ is the estimate of the standard deviation of the residuals.
- $ h_{ii} $ is the leverage of the $ i $-th observation, which measures the influence of the observation on the fitted values.

### Identifying Outliers Using Studentized Residuals

A common rule of thumb is that an observation is considered an outlier if the absolute value of its studentized residual is greater than 3. This threshold comes from the fact that in a normally distributed dataset, about 99.7% of the data should fall within three standard deviations from the mean. Therefore, a studentized residual beyond this range suggests that the observation is unusually far from what the model predicts.

### Plotting Studentized Residuals vs. Index

Now, let's create the plot of studentized residuals versus the index of observations. We will use a threshold of 3 and annotate any potential outliers.


```python
# Calculate studentized residuals
influence = model.get_influence()
studentized_residuals = influence.resid_studentized_internal

# Create a plot of studentized residuals vs index
plt.figure(figsize=(10, 6))
plt.plot(studentized_residuals, marker='o', linestyle='none', label='Studentized Residuals')
plt.axhline(y=3, color='red', linestyle='--', label='Threshold (3)')
plt.axhline(y=-3, color='red', linestyle='--')

# Annotate potential outliers
outliers = np.where(np.abs(studentized_residuals) > 3)[0]
for outlier in outliers:
    plt.annotate(outlier, (outlier, studentized_residuals[outlier]), textcoords="offset points", xytext=(0,10), ha='center')

plt.xlabel('Index')
plt.ylabel('Studentized Residuals')
plt.title('Studentized Residuals vs Index')
plt.legend()
plt.show()

outliers
```


### Interpretation of the Plot

The plot displays the studentized residuals for each observation in the dataset against their index. The red dashed lines at +3 and -3 represent the threshold for identifying outliers.

- **Outliers**: Observations with studentized residuals beyond the threshold (either above 3 or below -3) are considered potential outliers. In this case, observations at indices 1 and 2 have studentized residuals exceeding the threshold, making them potential outliers.

### Where Does the Threshold of 3 Come From?

The threshold of 3 is based on the properties of a standard normal distribution. For a normally distributed dataset, approximately 99.7% of the data should lie within three standard deviations of the mean. Therefore, studentized residuals greater than 3 (in absolute value) are rare and suggest that the corresponding observations might be outliers.

### Should Regression Diagnostics Be Repeated After Removing Potential Outliers?

Yes, regression diagnostics should be repeated after removing potential outliers. Removing outliers can significantly change the results of the regression analysis, including the estimated coefficients, residuals, and the fit of the model. By re-running the diagnostics, you can check whether the assumptions of the regression model are better satisfied and whether the overall fit of the model has improved. This iterative process helps ensure that the final model is robust and reliable.

## **High-Levergage Points**

### Understanding High-Leverage Points

**High-leverage points** are observations in a dataset that have an unusually large impact on the fit of a regression model. Leverage is a measure of how far an independent variable's value is from the mean of the independent variable values. An observation with high leverage has the potential to influence the estimated regression coefficients significantly, even if it is not an outlier in the dependent variable.

The leverage of an observation is measured by the diagonal elements of the hat matrix $ H $, denoted as $ h_{ii} $. The hat matrix maps the observed values to the fitted values.

The threshold for identifying high-leverage points is often set as $ \frac{2p}{n} $, where:
- $ p $ is the number of predictors (including the intercept) in the model.
- $ n $ is the total number of observations.

Observations with leverage values greater than $ \frac{2p}{n} $ are considered high-leverage points.

### Understanding Cook's Distance

**Cook's Distance** is a measure used to identify influential observations in a regression model. It combines information about the residual (how far the observed value is from the predicted value) and the leverage (how much influence the observation has on the model's fit). 

The basic idea behind Cook's Distance is that it measures how much the predicted values would change if a particular observation were removed from the dataset. A large Cook's Distance indicates that the observation has a strong influence on the fitted model.

#### Mathematical Equation

Cook's Distance for the $ i $-th observation is calculated as:

$
D_i = \frac{r_i^2}{p \hat{\sigma}^2} \times \frac{h_{ii}}{(1 - h_{ii})^2}
$

Where:
- $ r_i $ is the residual for the $ i $-th observation.
- $ h_{ii} $ is the leverage of the $ i $-th observation.
- $ p $ is the number of predictors in the model (including the intercept).
- $ \hat{\sigma}^2 $ is the estimated variance of the error term.

A common threshold for identifying influential points using Cook's Distance is $ \frac{4}{n} $. Observations with Cook's Distance greater than this threshold are considered potentially influential.

### Identifying High-Leverage Points and Influential Observations

Now, let's plot the leverage vs. index using the $ \frac{2p}{n} $ threshold and the Cook's Distance vs. index using the $ \frac{4}{n} $ threshold. We'll annotate observations that exceed these thresholds.

#### Plotting Leverage vs. Index


```python
# Calculate leverage (hat values) and Cook's Distance
leverage = influence.hat_matrix_diag
cooks_d = influence.cooks_distance[0]

# Number of predictors including intercept
p = X.shape[1]
n = len(X)

# Thresholds
leverage_threshold = 2 * p / n
cooks_d_threshold = 4 / n

# Plotting leverage vs. index
plt.figure(figsize=(10, 6))
plt.stem(range(n), leverage, linefmt='k-', markerfmt='ko', basefmt=" ", label='Leverage')
plt.axhline(y=leverage_threshold, color='red', linestyle='--', label=f'Threshold ({leverage_threshold:.3f})')

# Annotate high-leverage points
high_leverage_points = np.where(leverage > leverage_threshold)[0]
plt.stem(high_leverage_points, leverage[high_leverage_points], linefmt='r-', markerfmt='ro', basefmt=" ")

for point in high_leverage_points:
    plt.annotate(point, (point, leverage[point]), textcoords="offset points", xytext=(0,10), ha='center')

plt.xlabel('Index')
plt.ylabel('Leverage')
plt.title('Leverage vs Index')
plt.legend()
plt.show()

# Plotting Cook's Distance vs. index
plt.figure(figsize=(10, 6))
plt.stem(range(n), cooks_d, linefmt='k-', markerfmt='ko', basefmt=" ", label="Cook's Distance")
plt.axhline(y=cooks_d_threshold, color='red', linestyle='--', label=f'Threshold ({cooks_d_threshold:.3f})')

# Annotate points above Cook's Distance threshold
high_cooks_points = np.where(cooks_d > cooks_d_threshold)[0]
plt.stem(high_cooks_points, cooks_d[high_cooks_points], linefmt='r-', markerfmt='ro', basefmt=" ")

for point in high_cooks_points:
    plt.annotate(point, (point, cooks_d[point]), textcoords="offset points", xytext=(0,10), ha='center')

plt.xlabel('Index')
plt.ylabel("Cook's Distance")
plt.title("Cook's Distance vs Index")
plt.legend()
plt.show()

high_leverage_points, high_cooks_points
```



### Interpretation of the Plots

#### Leverage vs. Index Plot

## **Non-Linearity**

### Understanding Non-Linearity

**Non-linearity** refers to a situation in regression analysis where the relationship between the independent variables ($x_{ij}$) and the dependent variable ($y_i$) is not linear. In a linear regression model, the assumption is that the effect of the independent variables on the dependent variable is linear, meaning that the change in the dependent variable is proportional to the change in the independent variables.

If the true relationship between the variables is non-linear, but a linear model is applied, the model may produce biased estimates, poor predictions, and incorrect conclusions. Non-linearity can be identified by examining the residuals (the differences between observed and predicted values). If the residuals show a pattern when plotted against the fitted values, it indicates that the linear model might not be appropriate, and a non-linear model may be more suitable.

### Understanding the Rainbow Test

The **Rainbow Test** is a statistical test used to detect non-linearity in a regression model. It works by splitting the dataset into two partsâ€”typically the central part and the tails (beginning and end) of the dataset. The test then compares the fit of the model to the central portion of the data versus the fit to the entire dataset.

#### Basic Idea

The basic idea of the Rainbow Test is that if the model fits well to the central portion of the data but not to the tails, it suggests that the relationship between the variables might not be linear across the entire range of data. The test statistic compares the residual sum of squares (RSS) of the model when applied to the central part versus the whole dataset.

#### Mathematical Equations

Let:
- $ RSS_{\text{central}} $ be the residual sum of squares for the central portion of the data.
- $ RSS_{\text{full}} $ be the residual sum of squares for the entire dataset.

The Rainbow Test statistic is calculated as:

$
\text{Rainbow Test Statistic} = \frac{RSS_{\text{central}} / (n_{\text{central}} - p)}{RSS_{\text{full}} / (n - p)}
$

Where:
- $ n_{\text{central}} $ is the number of observations in the central portion.
- $ n $ is the total number of observations.
- $ p $ is the number of predictors in the model.

This statistic follows an F-distribution, and a high value indicates a potential violation of linearity.

### Conducting the Rainbow Test and Plotting Residuals vs. Fitted Values

Let's perform the Rainbow Test on the regression model and create a plot of residuals vs. fitted values to visually assess non-linearity.


```python
from statsmodels.stats.diagnostic import linear_rainbow

# Calculate predictions and residuals
predictions = model.fittedvalues
residuals = model.resid

# Perform the Rainbow Test
rainbow_statistic, rainbow_p_value = linear_rainbow(model)

# Plot residuals vs fitted values
plt.figure(figsize=(10, 6))
sns.residplot(x=predictions, y=residuals, lowess=True, line_kws={'color': 'red'})
plt.scatter(predictions, residuals, color='black')
plt.axhline(y=0, color='gray', linestyle='--')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.title('Residuals vs Fitted Values')
plt.show()

rainbow_statistic, rainbow_p_value
```


### Interpretation of the Rainbow Test and Residuals vs. Fitted Values Plot

#### Rainbow Test Results:

- **P-Value**: The p-value is extremely small (close to $2.42 \times 10^{-20}$).

Given that the p-value is much smaller than typical significance levels (e.g., 0.05), we reject the null hypothesis of linearity. This suggests that there may be non-linearity in the relationship between the independent variables and the dependent variable in the model.

#### Residuals vs. Fitted Values Plot:
- **Red Line (Lowess Smoothing)**: The red line in the plot represents a locally weighted scatterplot smoothing (Lowess) curve that fits the relationship between the residuals and the fitted values.
- **Scatter of Residuals**: The black dots represent the residuals plotted against the fitted values.

If the relationship between the independent and dependent variables were perfectly linear, we would expect the residuals to be randomly scattered around the horizontal line at zero, without any discernible pattern. However, if the residuals display a pattern or curvature, it suggests that the model may not be capturing the underlying relationship accurately and that non-linearity may be present.

### Interpretation:

Given these findings, further investigation and potentially adjusting the model to account for non-linear relationships may be warranted.

## **Heteroscedasticity**

### Understanding Heteroscedasticity

**Heteroscedasticity** refers to a situation in regression analysis where the variance of the errors (residuals) is not constant across all levels of the independent variables. In other words, the spread of the residuals changes depending on the value of the independent variables or the fitted values. This is a violation of one of the key assumptions of the classical linear regression model, which assumes homoscedasticityâ€”meaning that the error variance is constant.

When heteroscedasticity is present, it can lead to inefficient estimates of the regression coefficients and may result in misleading statistical inferences. Specifically, heteroscedasticity can affect the accuracy of confidence intervals and hypothesis tests, making it harder to trust the results of the regression analysis.

### Understanding the Breusch-Pagan Test

The **Breusch-Pagan Test** is a statistical test used to detect the presence of heteroscedasticity in a regression model. The test works by examining whether the variance of the residuals from a regression model depends on the independent variables.

#### Basic Idea

The basic idea behind the Breusch-Pagan Test is to regress the squared residuals from the original regression model on the independent variables (or their functions, like squares or interactions) to see if there is a systematic pattern. If the test finds a significant relationship between the residual variance and the independent variables, it suggests that heteroscedasticity is present.

#### Mathematical Equations

Let:
- $ r_i^2 $ be the squared residuals from the original regression model.
- $ X $ be the matrix of independent variables.

The Breusch-Pagan Test involves the following steps:

1. **Estimate the original regression model** and obtain the residuals $ r_i $.
2. **Calculate the squared residuals** $ r_i^2 $.
3. **Regress $ r_i^2 $ on the independent variables** $ X $ from the original model:

   $
   r_i^2 = \alpha_0 + \alpha_1 x_{i1} + \alpha_2 x_{i2} + \dots + \alpha_k x_{ik} + u_i
   $

4. **Compute the test statistic**:

   $
   LM = \frac{n R^2}{2}
   $

   where:
   - $ n $ is the number of observations.
   - $ R^2 $ is the coefficient of determination from the regression of $ r_i^2 $ on $ X $.

5. **Compare the test statistic** to a chi-square distribution with $ k $ degrees of freedom (where $ k $ is the number of independent variables). A significant test statistic indicates the presence of heteroscedasticity.

### Conducting the Breusch-Pagan Test and Creating the Scale-Location Plot

Let's conduct the Breusch-Pagan Test to check for heteroscedasticity and create the Scale-Location plot, which is another way to visually assess the presence of heteroscedasticity.


```python
from statsmodels.stats.diagnostic import het_breuschpagan
from statsmodels.stats.outliers_influence import OLSInfluence

# Perform the Breusch-Pagan test
bp_test = het_breuschpagan(residuals, model.model.exog)

# Extract the test statistic and p-value
bp_statistic, bp_p_value = bp_test[0], bp_test[1]

# Calculate the standardized residuals and their square root
standardized_residuals = OLSInfluence(model).resid_studentized_internal
sqrt_standardized_residuals = np.sqrt(np.abs(standardized_residuals))

# Create the Scale-Location plot
plt.figure(figsize=(10, 6))
sns.regplot(x=predictions, y=sqrt_standardized_residuals, lowess=True, line_kws={'color': 'red'})
plt.scatter(predictions, sqrt_standardized_residuals, color='black')
plt.xlabel('Fitted Values')
plt.ylabel('Sqrt(Standardized Residuals)')
plt.title('Scale-Location Plot')
plt.show()

bp_statistic, bp_p_value
```


### Scale-Location Plot

A **Scale-Location Plot** (also known as a Spread-Location Plot) shows the square root of the standardized residuals against the fitted values. If the plot shows a horizontal line with randomly scattered points, it indicates homoscedasticity. A clear pattern, such as a funnel shape, suggests heteroscedasticity.

Let's perform the Breusch-Pagan Test and create the plot.

### Interpretation of the Breusch-Pagan Test and Scale-Location Plot

#### Breusch-Pagan Test Results:
- **Test Statistic**: The Breusch-Pagan Test statistic is approximately 33.49.
- **P-Value**: The p-value is very small, around $ 2.54 \times 10^{-7} $.

Given the extremely low p-value, we reject the null hypothesis of homoscedasticity. This result strongly suggests that heteroscedasticity is present in the model.

#### Scale-Location Plot:

If the assumption of homoscedasticity holds, we would expect the plot to show a random scatter of points around a horizontal line, indicating that the variance of the residuals is constant across all levels of the fitted values.

However, if there is a clear pattern or trend in the plot, such as a funnel shape (where the residuals fan out as the fitted values increase), it suggests that the variance of the residuals is not constant, indicating heteroscedasticity.

### Interpretation:

## **Correlation of Error Terms**

### Understanding Correlation of Error Terms

**Correlation of error terms** refers to a situation where the residuals (errors) in a regression model are not independent of each other. This is also known as **autocorrelation**. Ideally, the residuals in a regression model should be uncorrelated, meaning the error for one observation should not predict or influence the error for another observation. When errors are correlated, it suggests that there is some underlying structure in the data that the model has not captured, which can lead to inefficiencies in estimating the regression coefficients and can invalidate standard hypothesis tests.

Autocorrelation is particularly common in time series data, where the value of one observation is often related to previous values.

### Understanding the Durbin-Watson Test

The **Durbin-Watson Test** is a statistical test used to detect the presence of autocorrelation in the residuals of a regression model, specifically first-order autocorrelation. The test calculates a statistic that ranges from 0 to 4, where:
- A value of **2** indicates no autocorrelation.
- A value close to **0** suggests positive autocorrelation (successive residuals are positively correlated).
- A value close to **4** suggests negative autocorrelation (successive residuals are negatively correlated).

#### Basic Idea

The basic idea of the Durbin-Watson Test is to compare the differences between consecutive residuals. If these differences are small, it suggests that the residuals are similar and hence correlated. The test provides a way to quantify this relationship and determine whether autocorrelation is present.

#### Mathematical Equations

The Durbin-Watson statistic $ d $ is calculated as:

$
d = \frac{\sum_{i=2}^{n} (e_i - e_{i-1})^2}{\sum_{i=1}^{n} e_i^2}
$

Where:
- $ e_i $ is the residual for the $ i $-th observation.
- $ n $ is the total number of observations.

The statistic $ d $ ranges from 0 to 4:
- **d â‰ˆ 2**: No autocorrelation.
- **d < 2**: Positive autocorrelation.
- **d > 2**: Negative autocorrelation.

A commonly recommended range for $ d $ to indicate no significant autocorrelation is between **1.5 and 2.5**.

### Conducting the Durbin-Watson Test and Plotting Residuals Over Time

Let's conduct the Durbin-Watson Test on the regression model to check for autocorrelation of the error terms and create a plot of the residuals over time (or index) to visually assess this correlation.


### Interpretation of the Durbin-Watson Test and Residuals Over Time Plot

#### Durbin-Watson Test Results:
- **Durbin-Watson Statistic**: The Durbin-Watson statistic is approximately 1.02.

Given the recommended range of 1.5 to 2.5 for indicating no significant autocorrelation, a Durbin-Watson statistic of 1.02 suggests the presence of positive autocorrelation in the residuals. This means that the residuals are likely correlated with each other, which violates the assumption of independent errors in the regression model.

#### Plot of Residuals Over Time:

### Interpretation:

This indicates that the residuals are not independent and that the model may not have captured all the underlying patterns in the data. This violation can lead to inefficiencies in estimating the regression coefficients and can affect the validity of statistical inferences, such as confidence intervals and hypothesis tests.

Given these findings, further steps might be necessary to address the autocorrelation issue, such as modifying the model to include lagged variables or using generalized least squares.

## **Normality of Residuals**

### Understanding Normality of Residuals

**Normality of residuals** is an assumption in linear regression that the residuals (the differences between the observed and predicted values) should be normally distributed. This assumption is important because many statistical tests and confidence intervals rely on the normality of residuals. If the residuals are normally distributed, it indicates that the model is appropriately capturing the underlying patterns in the data. If the residuals deviate significantly from normality, it can suggest that the model is missing key relationships, and the validity of statistical inferences (such as p-values and confidence intervals) may be compromised.

### Understanding the Shapiro-Wilk Test

The **Shapiro-Wilk Test** is a statistical test used to assess the normality of a dataset. Specifically, it tests the null hypothesis that the data comes from a normally distributed population. It is one of the most powerful tests for detecting deviations from normality.

#### Basic Idea

The basic idea behind the Shapiro-Wilk Test is to compare the order statistics (i.e., the sorted values) of the data with the expected order statistics if the data were normally distributed. The test calculates a statistic that measures how close the sample distribution is to a normal distribution.

#### Mathematical Equations

Let:
- $ x_1, x_2, \dots, x_n $ be the ordered sample data (i.e., sorted from smallest to largest).
- $ m_1, m_2, \dots, m_n $ be the expected values of the order statistics from a standard normal distribution.

The Shapiro-Wilk statistic $ W $ is calculated as:

$
W = \frac{\left( \sum_{i=1}^{n} a_i x_i \right)^2}{\sum_{i=1}^{n} (x_i - \bar{x})^2}
$

Where:
- $ a_i $ are constants calculated based on the expected order statistics $ m_i $.
- $ \bar{x} $ is the mean of the sample data.

The test statistic $ W $ ranges from 0 to 1:
- A value close to 1 indicates that the sample data is close to normal.
- A smaller value suggests deviation from normality.

The p-value associated with the test statistic helps determine whether to reject the null hypothesis of normality. If the p-value is less than a chosen significance level (e.g., 0.05), it suggests that the data is not normally distributed.

### Conducting the Shapiro-Wilk Test and Creating the QQ Plot

Let's perform the Shapiro-Wilk Test on the residuals from the regression model and create a QQ plot of the standardized residuals to visually assess their normality.


```python
from scipy.stats import shapiro

# Perform the Shapiro-Wilk test
shapiro_statistic, shapiro_p_value = shapiro(standardized_residuals)

# Create the QQ plot
plt.figure(figsize=(10, 6))
stats.probplot(standardized_residuals, dist="norm", plot=plt)
plt.title('QQ Plot of Standardized Residuals')
plt.show()

shapiro_statistic, shapiro_p_value
```


### QQ Plot

A **QQ Plot** (Quantile-Quantile Plot) compares the quantiles of the residuals with the quantiles of a normal distribution. If the residuals are normally distributed, the points should lie approximately on a straight line. Deviations from the line indicate departures from normality.

### Interpretation of the Shapiro-Wilk Test and QQ Plot

- **P-Value**: The p-value is very small, around $4.55 \times 10^{-8}$.

Given that the p-value is significantly lower than typical significance levels (e.g., 0.05), we reject the null hypothesis that the residuals are normally distributed. This result suggests that the residuals do not follow a normal distribution, indicating a violation of the normality assumption.

#### QQ Plot of Standardized Residuals:

  - Deviations from this line indicate departures from normality. In the plot, if the points deviate significantly from the line, especially at the tails, it suggests that the residuals are not normally distributed.

### Conclusion:
- **Non-Normality of Residuals Detected**: Both the Shapiro-Wilk test results and the QQ plot suggest that the residuals do not follow a normal distribution. This violation of the normality assumption can affect the validity of statistical tests and confidence intervals based on the model. If normality is crucial for your analysis, you may need to consider transforming the data, using a different modeling approach, or applying robust statistical methods that do not rely on the normality assumption.

## **Collinearity of Predictors**

### Understanding Collinearity of Predictors

**Collinearity** (or multicollinearity) refers to a situation in which two or more predictor variables in a regression model are highly correlated, meaning that they contain similar information about the variance in the dependent variable. When predictors are highly collinear, it becomes difficult to separate their individual effects on the dependent variable. This can lead to instability in the estimation of regression coefficients, making them unreliable and inflating the standard errors. In extreme cases, it might be impossible to determine the unique contribution of each predictor to the model.

### Understanding the Variance Inflation Factor (VIF)

The **Variance Inflation Factor (VIF)** is a measure used to detect the presence of multicollinearity in a regression model. It quantifies how much the variance of a regression coefficient is inflated due to collinearity with other predictors.

#### Basic Idea

The basic idea behind VIF is to assess how much a particular predictor variable is explained by the other predictors in the model. If a predictor is highly collinear with other variables, its VIF will be high, indicating that its variance is inflated due to multicollinearity.

#### Mathematical Equations

For each predictor $ X_j $, VIF is calculated as:

$
\text{VIF}_j = \frac{1}{1 - R_j^2}
$

Where:
- $ R_j^2 $ is the coefficient of determination obtained by regressing the $ j $-th predictor on all the other predictors in the model.

A high $ R_j^2 $ (close to 1) means that the predictor is highly correlated with the other predictors, leading to a high VIF value. A common rule of thumb is that a VIF value greater than 10 indicates significant multicollinearity that could be problematic.

### Conducting the Variance Inflation Factor (VIF) Analysis and Correlation Matrix Plot

Let's calculate the VIF for each predictor to check for multicollinearity. Additionally, we'll create a correlation matrix plot to visualize the correlations between the predictors.
